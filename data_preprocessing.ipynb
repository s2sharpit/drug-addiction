```python
# Data Loading and Preprocessing

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Dataset
train_df = pd.read_csv("train_drop_records.csv")

# Prepare features and target
X = train_df.drop("Addiction_Class", axis=1)
y = train_df["Addiction_Class"]
```

```python
# --- Correlation Analysis ---
plt.figure(figsize=(12, 10))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.tight_layout()
plt.savefig("correlation_matrix.png", dpi=300)
plt.show()
```

```python
# Drop highly correlated features
def drop_high_corr_features(data, threshold=0.9):
    corr_matrix = data.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return data.drop(columns=to_drop), to_drop

X, dropped_features = drop_high_corr_features(X)
print("Dropped highly correlated features:", dropped_features)
```

```python
# Save the cleaned data for feature selection and model training
X.to_csv("cleaned_features.csv", index=False)
y.to_csv("target.csv", index=False)
```
